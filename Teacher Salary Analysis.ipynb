```python
def rename_caps(df):
    return df.rename(columns=str.lower)
odf10,odf11,odf12,odf13,odf14,odf15,odf16,odf17,odf18= [rename_caps(df) for df in (odf10,odf11,odf12,odf13,odf14,odf15,odf16,odf17,odf18)]


def filter_jobs(df): 
    return df.loc[df['occ_code'].isin(codes)]

odf02,odf03,odf04,odf05,odf06,odf07,odf08,odf09,odf10,odf11,odf12,odf13,odf14,odf15,odf16,odf17,odf18,odf19= [filter_jobs(df) for df in (odf02,odf03,odf04,odf05,odf06,odf07,odf08,odf09,odf10,odf11,odf12,odf13,odf14,odf15,odf16,odf17,odf18,odf19)]
```

```python
def fix_w(df):
    return df.rename(columns={'a_wpct10':'a_pct10','a_wpct90':'a_pct90','h_wpct10':'h_pct10','h_wpct90':'h_pct90',})
odf01,odf02 = [fix_w(df)for df in (odf01,odf02)]
```

```python
odfs = [odf01,odf02,odf03,odf04,odf05,odf06,odf07,odf08,odf09,odf10,odf11,odf12,odf13,odf14,odf15,odf16,odf17,odf18,odf19]

odfAll = pd.concat(odfs, sort=True)
```

```python
odfAll.head()
```

```python
odfAll = odfAll.drop(columns=['group','o_group','occ_group'])
odfAll
```

```python
odfAll[['a_median','a_pct10','a_pct90']] = odfAll[['a_median','a_pct10','a_pct90']].apply(pd.to_numeric,errors='coerce')

```

```python
odfAll = odfAll.dropna(subset=['a_median','a_pct10','a_pct90'])
```

```python
myMap ={'13-1071':'HR Specialists', 
        '13-2010':'Accountants',
        '13-2081':'Tax Examiners', 
        '21-1022':'Social Workers', 
        '33-2010':'Firefighters',
        '33-3051':'Police Officers',
        '43-5052':'Mail Carriers', 
        '25-2031':'HS Teachers'}

tempJobs = odfAll['occ_code'].map(myMap)
odfAll['job'] = tempJobs
```

```python
medianAnnuals = odfAll.groupby(['year', 'job'])['a_median'].mean()
medianAnnuals.unstack().plot()
```
