Next we clean our data: 
1. Fixing the naming convention for counties 

```python
#in remove rows for Arizona/New Jersey in general
incomesAZ.drop(incomesAZ.head(3).index, inplace = True)
incomesNJ.drop(incomesNJ.head(3).index, inplace = True)
```

```python
incomesAZ['CountyList'] = incomesAZ['GeoName'].str.split(',')
incomesAZ.loc[:, 'County'] = incomesAZ.CountyList.map(lambda x: x[0])
incomesAZ.head()
```

```python
incomesNJ['CountyList'] = incomesNJ['GeoName'].str.split(',')
incomesNJ.loc[:, 'County'] = incomesNJ.CountyList.map(lambda x: x[0])
```
